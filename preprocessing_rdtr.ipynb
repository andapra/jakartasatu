```python
domains = arcpy.da.ListDomains(os.path.join(workspace, 'new_rtrw_rdtr', '2.Pola Ruang RDTR', '_3100_RDTRDKIJAKARTA_2022.gdb'))
list_domain_rdtr = {}
for domain in domains:
    get_domain_name = domain.name
    if domain.domainType == 'CodedValue':
        coded_values = domain.codedValues
        list_domain_rdtr['{}'.format(get_domain_name)] = coded_values
```

```python
list_domain_rdtr.keys()
```

```python
list_conv_str = ['KODZON', 'JNSRPR', 'KODSBL', 'KKOP', 'LP2B', 'KRB', 'TOD', 'TEB', 'PUSLIT', 'CAGBUD',
       'RESAIR', 'KSMPDN', 'HANKAM', 'KKARST', 'PTBGMB', 'MGRSAT', 'RDBUMI']
```

```python
for subindex in list_conv_str:
    for index, row in sdf_rdtr.iterrows():
        get_column_value = row['{}'.format(subindex)]
        get_domain = list_domain_rdtr['{}'.format(subindex)]
        if get_column_value is not None:
            if get_column_value in get_domain:
                get_subvalue = get_domain[get_column_value]
                sdf_rdtr.loc[index, '{}_conv'.format(subindex)] = get_subvalue
        
```
